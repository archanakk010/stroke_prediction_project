```python
df["Channel"] = df["Channel"].str.lower().str.strip()
```

```python
df["Channel"].value_counts()
```

```python
df["Channel"] = df["Channel"].replace({
    
    # Email
    "email": "Email",
    
    # Social Media
    "social media": "Social Media",
    "socialmedia": "Social Media",
    "social-media": "Social Media",
    
    # Paid Ads
    "paidads": "Paid Ads",
    "paid ads": "Paid Ads",
    
    # Affiliate
    "affiliate": "Affiliate",
    "affilate": "Affiliate",
    "affilate": "Affiliate",
    
    # SEO
    "seo": "SEO"
})
```

```python
df["Channel"].value_counts()
```

```python
df["Target_Age_Group"].value_counts()
```

```python
df["Target_Age_Group"] = df["Target_Age_Group"].str.replace("_","-")
df["Target_Age_Group"] = df["Target_Age_Group"].str.replace("–","-")
df["Target_Age_Group"] = df["Target_Age_Group"].str.replace(" ","")
```

```python
df["Target_Age_Group"].value_counts()
```

```python
df["Start_Date"] = pd.to_datetime(df["Start_Date"], errors="coerce")
df["End_Date"] = pd.to_datetime(df["End_Date"], errors="coerce")
```

```python
df.info()
```

```python
df["Campaign_Duration_days"] = (df["End_Date"] - df["Start_Date"]).dt.days

df.head()
```

```python
df[df["Campaign_Duration_days"] < 0]
```

```python
len(df[df["Campaign_Duration_days"] < 0])
```

```python
df.loc[df["Campaign_Duration_days"] < 0, "Campaign_Duration_days"] = np.nan
```

```python

df["CPC"] = np.where(df["Clicks"] > 0,
                     df["Budget"] / df["Clicks"],
                     np.nan)
df
```

```python
df["CPA"] = np.where(df["Conversions"] > 0,
                     df["Budget"] / df["Conversions"],
                     np.nan)
df
```

```python
df["ROI"] = (df["Revenue"] - df["Budget"]) / df["Budget"] * 100
df
```

```python
roi_channel = df.groupby("Channel")["ROI"].mean().reset_index()

roi_channel
```
